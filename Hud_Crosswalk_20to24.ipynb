```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')
```

```python
#import each hud crosswalk file

df_20 = pd.read_excel('/content/drive/MyDrive/HUD_Crosswalk/ZIP_TRACT_122020.xlsx')

df_21 = pd.read_excel('/content/drive/MyDrive/HUD_Crosswalk/ZIP_TRACT_122021.xlsx')

df_22 = pd.read_excel('/content/drive/MyDrive/HUD_Crosswalk/ZIP_TRACT_122022.xlsx')

df_23 = pd.read_excel('/content/drive/MyDrive/HUD_Crosswalk/ZIP_TRACT_122023.xlsx')

df_24 = pd.read_excel('/content/drive/MyDrive/HUD_Crosswalk/ZIP_TRACT_122024.xlsx')
```

```python
df_20['Year'] = 2020
df_21['Year'] = 2021
df_22['Year'] = 2022
df_23['Year'] = 2023
df_24['Year'] = 2024
```

```python
df_21 = df_21.rename(columns={
    'zip': 'ZIP',
    'tract': 'TRACT',
    'res_ratio': 'RES_RATIO',
    'bus_ratio': 'BUS_RATIO',
    'oth_ratio': 'OTH_RATIO',
    'tot_ratio': 'TOT_RATIO',
    'usps_zip_pref_city': 'USPS_ZIP_PREF_CITY',
    'usps_zip_pref_state': 'USPS_ZIP_PREF_STATE'
})
```

```python
df_hud = pd.concat([df_20, df_21, df_22, df_23, df_24], ignore_index=True)

df_hud.info()
```

```python
df_hud = df_hud.drop(columns=['USPS_ZIP_PREF_CITY','USPS_ZIP_PREF_STATE'])
df_hud.info()
```

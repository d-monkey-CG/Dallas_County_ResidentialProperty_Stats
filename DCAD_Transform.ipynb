```python
df_2020 = pd.read_csv('/content/drive/MyDrive/Dallas_County_25YR_Data/DCAD2020_CURRENT/DCAD_Joined_2020.csv', low_memory=False)
df_2021 = pd.read_csv('/content/drive/MyDrive/Dallas_County_25YR_Data/DCAD2021_CURRENT/DCAD_Joined_2021.csv', low_memory=False)
df_2022 = pd.read_csv('/content/drive/MyDrive/Dallas_County_25YR_Data/DCAD2022_CURRENT/DCAD_Joined_2022.csv', low_memory=False)
df_2023 = pd.read_csv('/content/drive/MyDrive/Dallas_County_25YR_Data/DCAD2023_CURRENT/DCAD_Joined_2023.csv', low_memory=False)
df_2024 = pd.read_csv('/content/drive/MyDrive/Dallas_County_25YR_Data/DCAD2024_CURRENT/DCAD_Joined_2024.csv', low_memory=False)

df_combined = pd.concat([df_2020, df_2021, df_2022, df_2023, df_2024], ignore_index=True)

df_combined.info()
```

```python
#To check zipcode columns for errors

df_combined[df_combined['PROPERTY_ZIPCODE'].isna()]
```

```python
df_combined.info()
```

```python
#To change the POOL_IND datatype from object to boolean

df_combined['POOL_IND'] = df_combined['POOL_IND'].astype(bool)
```

```python
df_combined.info()
```

```python
df_combined.head(15)
```

```python
#To drop all rows where the YR_BUILT column has null or zero values

df_combined = df_combined.dropna(subset=['YR_BUILT'])

df_combined = df_combined[df_combined['YR_BUILT'] != 0]

df_combined.head(15)
```

```python
#To truncate the PROPERTY_ZIPCODE column values to 5 digit zipcodes only

df_combined['PROPERTY_ZIPCODE'] = df_combined['PROPERTY_ZIPCODE'].astype(str).str[:5]

#To drop rows with null values and convert the zipcode column to numeric

df_combined['PROPERTY_ZIPCODE'] = pd.to_numeric(
    df_combined['PROPERTY_ZIPCODE'], errors="coerce"
)

df_combined = df_combined.dropna(subset=['PROPERTY_ZIPCODE'])

df_combined['PROPERTY_ZIPCODE'] = df_combined['PROPERTY_ZIPCODE'].astype('int64')

df_combined.info()
```
